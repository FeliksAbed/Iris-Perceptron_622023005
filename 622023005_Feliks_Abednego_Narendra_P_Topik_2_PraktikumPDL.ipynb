```python
def min_max_normalize(data):
    min_vals = data[:, :4].min(axis=0)
    max_vals = data[:, :4].max(axis=0)
    normalized = (data[:, :4] - min_vals) / (max_vals - min_vals)
    return normalized

norm_setosa = min_max_normalize(iris_setosa)
norm_versicolor = min_max_normalize(iris_versicolor)
norm_virginica = min_max_normalize(iris_virginica)
```

```python
import pandas as pd

norm_setosa_df = pd.DataFrame(norm_setosa, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
norm_setosa_df.plot()
```

```python
import pandas as pd

norm_versicolor_df = pd.DataFrame(norm_versicolor, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
norm_versicolor_df.plot()
```

```python
import pandas as pd

norm_virginica_df = pd.DataFrame(norm_virginica, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
norm_virginica_df.plot()
```

```python
combined_data = np.vstack([
    np.hstack([norm_setosa, np.zeros((norm_setosa.shape[0], 1))]),
    np.hstack([norm_versicolor, np.ones((norm_versicolor.shape[0], 1))]),
    np.hstack([norm_virginica, np.full((norm_virginica.shape[0],1),2)])
])
```

```python
Feliks = pd.DataFrame(combined_data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])
Feliks.to_csv('Percobaan3_622023005.csv', index=False)
```
